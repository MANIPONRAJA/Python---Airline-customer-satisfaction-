```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('passenger.csv')
```

```python
pd.set_option('display.max_columns', 24)
pd.set_option('display.max_rows', 103904)
```

```python
df.head()
```

```python
# to calculate number of passengers based on age
plt.style.use('fivethirtyeight') 

ages = df['Age']
 
bins=[10,20,30,40,50,60,70,80,90]

plt.hist(ages, bins=bins, edgecolor='black', log=True) 

median_age = df['Age'].median()
color = '#fc4f30'

plt.axvline(median_age, color=color, label="Age Median")
plt.legend()

plt.title('Ages of Passengers')
plt.xlabel('Ages')
plt.ylabel('Total Passengers')

plt.tight_layout()

plt.show()

print('People who travel most are from age 40-50 due to comfort or maybe mostly business people')
```

```python
plt.style.use('seaborn')

Distance = df['Flight Distance']
Delay = df['Arrival Delay in Minutes']
Cleanliness = df['Cleanliness']

# 
plt.scatter(Distance, Delay, c=Cleanliness, cmap='summer',edgecolor='black',linewidth=1,alpha=0.75)

cbar=plt.colorbar()
cbar.set_label('Cleanliness')

plt.title('Passenger Satisfaction')
plt.xlabel('Distance(km)')
plt.ylabel('Delay(minutes)')

plt.tight_layout()

plt.show()

print('The fights which are travelling more distance are making less delays compare to others but in the shorter travel flight cleanliness are poor due to their repeated travel')
```

```python
plt.style.use("fivethirtyeight")

class_counts = df['Class'].value_counts()

plt.pie(class_counts, labels = class_counts.index, wedgeprops={'edgecolor': 'black'})

plt.title('Distribution of Travel Classes')
plt.tight_layout()
plt.show()
```
